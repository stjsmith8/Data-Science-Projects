## FIFA Descriptive Analysis

This notebook contains a descriptive analysis of FIFA soccer player stats.

```python
# load libraries
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
# Import file
fifa = pd.read_csv('players_20.csv')
```

```python
fifa.head()
```

```python
# Print columns (104 columns)
for col in fifa.columns:
    print(col)
```

```python
# Examine number of rows and columns
rows, columns = fifa.shape
print('Rows:', rows, 'Columns:', columns)
```

```python
# Examine counts of nationality
fifa['nationality'].value_counts()
```

```python
# Examine counts of nationality for 10 most common
fifa['nationality'].value_counts()[0:10]
```

```python
# Extract name of top 5 nationalities
fifa['nationality'].value_counts()[0:5].keys()
```

```python
# Create bar graph
plt.figure(figsize=(8,5))
plt.bar(
    list(fifa['nationality'].value_counts()[0:5].keys()),
    list(fifa['nationality'].value_counts()[0:5]),
    color="b")
plt.title("Bar Graph of Player Nationalities")
plt.xlabel("Country")
plt.ylabel("Count")
plt.show()
```

```python
# Create a new df
player_salary = fifa[['short_name', 'wage_eur']]
player_salary.head()
```

```python
# Sort df by ascending order
player_salary = player_salary.sort_values(by=['wage_eur'],ascending=False)
player_salary.head()
```

```python
# Graph top 5 earners
plt.figure(figsize=(8,5))
plt.bar(
    list(player_salary['short_name'])[0:5],
    list(player_salary['wage_eur'])[0:5],
    color=["red","blue", "green","orange","pink"])
plt.title("Bar Graph of Highest Earners")
plt.xlabel("Player Name")
plt.ylabel("Wage (in Euros)")
plt.show()
```

```python
# Create new df where nationality == Germany
Germany = fifa[fifa['nationality'] == 'Germany']
Germany.head()
```

```python
# Tallest German player
Germany.sort_values(by=['height_cm'],ascending=False).head()
```

```python
#Most paid German player
Germany.sort_values(by=['wage_eur'],ascending=False).head()
```

```python
# Find player with best shooting stat
player_shooting = fifa[['short_name', 'shooting']]
player_shooting.head()
```

```python
player_shooting.sort_values(by=['shooting'],ascending=False).head()
```

```python
# Find player with best defending stat
player_defending = fifa[['short_name', 'defending', 'nationality', 'club']].sort_values(by=['defending'],ascending=False)
player_defending.head()
```

```python
# Look at player stats for Real Madrid
real_madrid = fifa[fifa['club'] == 'Real Madrid']
print(real_madrid[['short_name', 'wage_eur']].sort_values(by=['wage_eur'],ascending=False).head())
print(real_madrid[['short_name','shooting']].sort_values(by=['shooting'],ascending=False).head())
print(real_madrid[['short_name', 'defending']].sort_values(by=['defending'],ascending=False).head())
```
